```python
import pandas as pd

# Load the dataset
data = pd.read_csv("EPF.csv") 

# Display the first few rows of the dataset to understand its structure
print(data.head())
```

```python
# Check for missing values
print(data.isnull().sum())
```

```python
# Basic statistics summary
print(data.describe())
```

```python
# EDA - Visualization (You can use libraries like matplotlib or seaborn)
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

# Example visualizations
# Distribution of age groups
plt.figure(figsize=(10, 6))
sns.countplot(x='age_group', data=data)
plt.title('Distribution of Age Groups')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()

```

```python
# Distribution of genders
plt.figure(figsize=(8, 5))
sns.countplot(x='gender', data=data)
plt.title('Distribution of Genders')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()


```

```python
# Relationship between age group and EPF subscriptions
plt.figure(figsize=(10, 6))
sns.boxplot(x='age_group', y='value', hue='category', data=data)
plt.title('Relationship between Age Group and EPF Subscriptions')
plt.xlabel('Age Group')
plt.ylabel('EPF Subscription Value')
plt.xticks(rotation=45)
plt.legend(title='Category')
plt.show()

# Conclusion:
# You can draw conclusions based on the visualizations and statistical summaries.
# For example, you can infer which age group has the highest EPF subscriptions or which gender dominates in EPF subscriptions.
# Further analysis can be done based on your specific requirements and questions of interest.
```

```python

# Year-wise analysis
yearly_data = data.groupby('fiscal_year').sum().reset_index()

# Calculate total subscribers after deducting exited members who rejoined and resubscribed
yearly_data['net_subscribers'] = yearly_data['Number of New EPF Subscribers during the year'] - yearly_data['Number of exited members who rejoined and resubscribed']

# Display the yearly analysis
print(yearly_data)

# EDA - Visualization for yearly analysis
import matplotlib.pyplot as plt

# Plotting year-wise net subscribers
plt.figure(figsize=(10, 6))
plt.plot(yearly_data['fiscal_year'], yearly_data['net_subscribers'], marker='o')
plt.title('Year-wise Net EPF Subscribers')
plt.xlabel('Fiscal Year')
plt.ylabel('Net Subscribers')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

# Conclusion:
# Based on the yearly analysis, you can observe trends in EPF subscriptions over the years and identify any significant changes or patterns.
```
